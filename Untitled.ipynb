```python
import os
import shutil
import pandas as pd

# Define paths
dataset_dir = 'dataset'
train_dir = os.path.join(dataset_dir, 'train')
test_dir = os.path.join(dataset_dir, 'test')
csv_file = 'Chest_xray_Corona_Metadata.csv'

# Read CSV file
df = pd.read_csv(csv_file)

# Iterate through each row
for index, row in df.iterrows():
    image_name = row['X_ray_image_name']
    label = row['Label']
    dataset_type = row['Dataset_type']

    if dataset_type == 'TRAIN':
        src_dir = train_dir
    elif dataset_type == 'TEST':
        src_dir = test_dir
    else:
        continue  # Skip if dataset type is neither TRAIN nor TEST

    # Check label and move image to appropriate folder
    if label == 'Normal':
        dest_dir = os.path.join(dataset_dir, 'normal')
    else:
        dest_dir = os.path.join(dataset_dir, 'pneumonia')

    # Make destination directory if it doesn't exist
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    # Move image to destination folder
    src_path = os.path.join(src_dir, image_name)
    dest_path = os.path.join(dest_dir, image_name)
    shutil.copy(src_path, dest_path)

    print(f"Moved '{image_name}' to '{dest_dir}'")
```

```python
import os
import cv2
import pandas as pd

# Define paths
dataset_dir = 'dataset'
normal_dir = os.path.join(dataset_dir, 'normal')
pneumonia_dir = os.path.join(dataset_dir, 'pneumonia')
normal_canny_dir = os.path.join('dataset_real', 'normal_canny')
pneumonia_canny_dir = os.path.join('dataset_real', 'pneumonia_canny')
csv_file = 'Chest_xray_Corona_Metadata.csv'

# Read CSV file
df = pd.read_csv(csv_file)

# Function to process and save image
def process_and_save_image(src_path, dest_dir):
    # Read the image
    img = cv2.imread(src_path,cv2.IMREAD_GRAYSCALE)

    # Get original image dimensions
    orig_h, orig_w = img.shape[:2]

    # Calculate new width and height (50% smaller)
    new_w = int(orig_w * 0.5)
    new_h = int(orig_h * 0.5)



    # Apply Canny edge detection
    edges = cv2.Canny(img, 40, 80)
        # Resize the image
    img_resized = cv2.resize(edges, (new_w, new_h))
    # Save the processed image
    dest_path = os.path.join(dest_dir, os.path.basename(src_path))
    cv2.imwrite(dest_path, img_resized)

# Create folders for processed images if they don't exist
for dir_path in [normal_canny_dir, pneumonia_canny_dir]:
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)

# Process and save normal images
for image_name in os.listdir(normal_dir):
    src_path = os.path.join(normal_dir, image_name)
    process_and_save_image(src_path, normal_canny_dir)
    print(f"Processed and saved '{image_name}' in '{normal_canny_dir}'")

# Process and save pneumonia images
for image_name in os.listdir(pneumonia_dir):
    src_path = os.path.join(pneumonia_dir, image_name)
    process_and_save_image(src_path, pneumonia_canny_dir)
    print(f"Processed and saved '{image_name}' in '{pneumonia_canny_dir}'")
```

```python
import streamlit as st
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Image dimensions
IMG_W = 100
IMG_H = 200
CHANNELS = 3  # Assuming RGB images

# Load the trained model
model = load_model('UASPCD.keras')

# Function to preprocess the uploaded image
def preprocess_image(image):
  # Read the image in color (ensure color mode is compatible with model)
  image = cv2.imdecode(np.frombuffer(uploaded_file.read(), np.uint8), cv2.IMREAD_COLOR)

  # Resize the image
  image = cv2.resize(image, (IMG_W, IMG_H))

  # Normalize the image (assuming model expects normalized values)
  image = image.astype('float32') / 255.0

  # Expand dimensions for batch processing (even for single image)
  image = np.expand_dims(image, axis=0)

  return image

# Function to make predictions using the loaded model
def predict_image(image):
  # Preprocess the image
  img_array = preprocess_image(image)

  # Make prediction (no data augmentation during prediction)
  prediction = model.predict(img_array)
  return prediction

# Streamlit app
def main():
  st.title("Image Prediction App")

  uploaded_file = st.file_uploader("Choose an image file", type=['jpg', 'jpeg', 'png'])

  if uploaded_file is not None:
    # Display the uploaded image
    image = cv2.imdecode(np.frombuffer(uploaded_file.read(), np.uint8), cv2.IMREAD_COLOR)
    st.image(image, channels='BGR', caption='Uploaded Image', use_column_width=True)

    # Preprocess the image for prediction (no grayscale conversion)
    img_array = preprocess_image(image)

    # Make prediction
    prediction = predict_image(img_array)
    # You can interpret the prediction based on your model's output layer
    st.write(f"Prediction: {prediction}")

if __name__ == '__main__':
  main()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Flatten, Dense, Dropout
from tensorflow.keras.layers import Activation
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Constants
IMG_W = 100  # Width of resized images
IMG_H = 200  # Height of resized images
CHANNELS = 3  # Assuming RGB images
NB_CLASSES = 2  # Number of classes (normal and pneumonia)
EPOCHS = 30
BATCH_SIZE = 6
DATASET_DIR = 'dataset_real'  # Directory containing the images

# Define model
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=(IMG_H, IMG_W, CHANNELS)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(Conv2D(256, (3, 3)))
model.add(Activation('relu'))

model.add(Conv2D(128, (3, 3)))
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (2, 2)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(32))
model.add(Dropout(0.25))
model.add(Dense(1))
model.add(Activation('sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

# Image data generators (adjust parameters as needed)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.3)  # Set color_mode to 'grayscale'
validation_datagen = ImageDataGenerator(
    rescale=1./255)
train_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_H, IMG_W),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training')

validation_generator = validation_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_H, IMG_W),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False,
    subset='validation')


# Training using fit
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE,
    epochs=EPOCHS)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
